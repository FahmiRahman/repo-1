```python
def number_type(number):
    try:
        num = float(number)
        if num.is_integer():
            return "Integer"
        else:
            return "Decimal"
    except ValueError:
        return "Invalid input"
    
number_1 = input("Enter the first number: ")
number_2 = input("Enter the second number: ")

num_1_type = number_type(number_1)
num_2_type = number_type(number_2)

print(f"The first number is: {num_1_type}")
print(f"The second number is: {num_2_type}")
```

```python
def calculate(num_1, num_2, operation):
    result = None

    if operation == 'addition':
        result = num_1 + num_2
    elif operation == 'subtraction':
        result = num_1 - num_2
    elif operation == 'multiplication':
        result = num_1 * num_2
    elif operation == 'division':
        if num_2 != 0:
            result = num_1 / num_2
        else:
            print("Error: Division by zero!")
    else:
        print("Error: Invalid operation!")

    return result    
                
number_1 = float(input{"Enter the first number: "})
number_2 = float(input{"Enter the second number: "})
op = input("Enter operation {addition, subtraction, multiplication, division}: ")

result = calculate(number_1, number_2, op)
if result is not None:
    print(f"Result of {op} {number_1} and {number_2} is: {result}")
```
